```python
from transformers import pipeline

```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
prompt = "Artificial Intelligence will change the future by"

output = generator(
    prompt,
    max_length=100,
    num_return_sequences=1
)

print(output[0]["generated_text"])
```

```python
user_prompt = input("Enter your prompt: ")

response = generator(
    user_prompt,
    max_length=150,
    num_return_sequences=1
)

print("\nGenerated Text:\n")
print(response[0]["generated_text"])
```
