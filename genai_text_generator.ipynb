```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
prompt = "Explain Ohm's Law in simple words."

output = generator(
    prompt,
    max_length=100,
    num_return_sequences=1
)

print(output[0]["generated_text"])
```

```python
while True:
    user_input = input("Enter prompt (or type exit): ")
    if user_input.lower() == "exit":
        break

    result = generator(user_input, max_length=120)
    print("\nAI:", result[0]["generated_text"])
```
